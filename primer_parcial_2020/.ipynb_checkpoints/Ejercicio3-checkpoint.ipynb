### Preprocesamiento

Se considerará a cada correo electrónico como una secuencia de palabras. Además del espacio, se consideran como separadores de palabras los siguientes caracteres: . , ? \t \n \r.  Se prové la función `obtener_lista_de_palabras_del_conjunto()` que realiza el *parseo*. Ejecutar la celda y continuar.

```python
def obtener_lista_de_palabras_del_conjunto(directorio):
    '''
    Parsea los correos del conjunto 
    
    Entrada:
        directorio: carpeta donde se encuentran los correos a parsear
    Salida:
        lista_palabras_conjunto: lista que almacena las palabras presentes en los correo. Cada lugar
                                 en la lista corresponde a un correo y contiene las palabras almacenadas
                                 en el correo.
    '''
    filenames = os.listdir(directorio)
    filenames.sort()
    lista_palabras_conjunto = []
    for i, filename in enumerate(filenames):
        if filename.endswith('.txt'):
            full_filename = os.path.join(directorio, filename)
            with open(full_filename, 'rb') as f:
                text = str(f.read()).lower()
                text = text.replace('.', ' ')
                text = text.replace('?', ' ')
                text = text.replace(',', ' ')
                text = text.replace('\\n', ' ')
                text = text.replace('\\t', ' ')
                text = text.replace('\\r', ' ')
                lista_palabras_correo = text.split(sep=' ')
                lista_palabras_correo = [w for w in lista_palabras_correo if w!='']
                lista_palabras_conjunto.append(lista_palabras_correo)
    return lista_palabras_conjunto
```

Los correos se codificarán utilizando la técnica *bag of words*. Dicha técnica codifica las palabras utilizando un diccionario. Se construirá un vector de características de largo igual a la cantidad de palabras del diccionario y cada componente del vector de características almacenará la cantidad de veces que aparece la palabra correspondiente en el correo. ### Ejercicio 3 - parte a)  

Completar la implementación de la función `codificar_conjunto()`. La misma codifica los *N* correos presentes en un conjunto. Cada correo se codifica con un vector de largo igual al número de palabras del diccionario (1014). Cada componente del vector es el número de veces que la palabra del diccionario aparece en el correo.

```python
def codificar_conjunto(correos, diccionario):
    '''
    Entrada:
        correos: lista de tamaño N que contiene los correos del conjunto
        diccionario: (K)
    Salida:
        caracteristicas: (N, K)
    '''
    N = len(correos)
    K = len(diccionario)
    caracteristicas = np.empty((N, K))
    
    ###########################################################
    ######  Empieza espacio para completar código  ############
    ###########################################################
    
    # en alguna implementación puede ser útil el método
    # de listas count(value).
    
    for i, correo in enumerate(correos):
        for k, palabra in enumerate(diccionario):
            caracteristicas[i, k] = correo.count(palabra)
    
    ###########################################################
    ######  Termina espacio para completar código  ############
    ###########################################################
        
    return caracteristicas
```

## Regresión Logística

Se utilizará *regresión logística* regularizada para entrenar el clasificador, para ello se minimiza la siguiente función de costo:
$$
E_{in} = \frac{1}{N}\sum_{i=1}^N \log \left( 1 + \exp \left( -y_n \mathbf{w}^T\mathbf{x}_i \right) \right) + \frac{\lambda}{2} \Vert \mathbf{w}\Vert_2^2 
$$

<a id="Parte4"></a>
### Ejercicio 3 - parte c)  

Mostrar que el gradiente de la función de costo puede escribirse como:

$$
\frac{\partial E_{in}}{\partial \mathbf{w}} = \frac{1}{N} \sum_{i=1}^{N} -y_i \mathbf{x}_i \theta \left( - y_i \mathbf{w}^T\mathbf{x_i} \right) + \lambda \mathbf{w}
$$

siendo $\theta \left( \mathbf{s} \right)$ la función *sigmoide* definida como $\theta \left( \mathbf{s} \right) = \frac{1}{1+\exp(-\mathbf{s})}$.

**Respuesta:**    

$$ \nabla_\textbf{w} E_{in} = \frac{1}{N} \sum_{i=1}^N \nabla_\textbf{w} \left[ log(1+\exp(-y_n \textbf{w}^T \textbf{x}_i)) + \frac{\lambda}{2} \textbf{w}^T \textbf{w} \right]$$

Evaluando el gradiente se tiene:

$$ \nabla_\textbf{w} E_{in} = \frac{1}{N} \sum_{i=1}^N -y_i \textbf{x}_i \frac{\exp(-y_n \textbf{w}^T \textbf{x}_i)}{1+\exp(-y_n \textbf{w}^T \textbf{x}_i)} + \lambda \textbf{w} = \frac{1}{N} \sum_{i=1}^N -y_i \textbf{x}_i \frac{1}{1+\exp(y_n \textbf{w}^T \textbf{x}_i)} + \lambda \textbf{w} $$

Observar que $\theta(-y_n \textbf{w}^T \textbf{x}_i) = \frac{1}{1+\exp(y_n \textbf{w}^T \textbf{x}_i)}$
Por lo tanto:
$$ \nabla_\textbf{w} E_{in} = \frac{1}{N} \sum_{i=1}^N -y_i \textbf{x}_i \theta(-y_n \textbf{w}^T \textbf{x}_i) + \lambda \textbf{w} $$

### Ejercicio 3 - parte d)  
Antes de implementar el clasificador de regresión logística se implementarán algunas funciones auxiliares. Implementar y verificar el correcto funcionamiento de las siguientes funciones: `sigmoide()`,  `coordenadas_homogeneas()`, `calcular_gradiente_y_costo_sin_regularizar()`.   


```python
def sigmoide(x):
    
    '''
    Entrada:
        x: arreglo al que se le aplicará la función sigmoide
    Salida:
        out: sigmoide(x)
    '''
    #################################################################################
    #################  EMPIEZA ESPACIO PARA COMPLETAR CÓDIGO ########################
    #################################################################################
    
    out = 1/(1+np.exp(-x))
    
    #################################################################################
    #################  EMPIEZA ESPACIO PARA COMPLETAR CÓDIGO ########################
    #################################################################################
    
    # Para evitar problemas numéricos se saturan los valores pequeños en 1e-16
    if np.isscalar(out):
        if out < 1e-16:
            out = 1e-16
    else:            
        out[out < 1e-16] =  1e-16

    return out
```

```python
# Se testea la función sigmoide
np.random.seed(32)
x5 = np.random.randn(3,2)
y5 = sigmoide(x5)
y5_ = np.array([[0.41365054, 0.72784244], [0.64127972, 0.51756388],[0.68514811, 0.64151799]])
np.allclose(y5, y5_)
```

```python
def coordenadas_homogeneas(X):
    '''
    Pasaje a coordenadas homogéneas
    '''
    
    #################################################################################
    #################  EMPIEZA ESPACIO PARA COMPLETAR CÓDIGO ########################
    #################################################################################
    
    M, d = X.shape
    unos = np.ones((M,1))
    Xh = np.hstack((unos,X))
    
    #################################################################################
    #################  TERMINA ESPACIO PARA COMPLETAR CÓDIGO ########################
    #################################################################################
    
    return Xh
```

```python
def calcular_costo_y_gradiente_sin_regularizar(X, y, w):
        '''
        Calcula el costo y el gradiente del costo utilizando todas las muestras 
        disponibles para el entrenamiento
        Entrada:
            X: arreglo de (N,d+1) que contiene las características de los datos de entrada. 
               Las características están expresadas en coordenadas homogéneas
            y: arreglo de largo N que contiene las etiquetas   
            w: vector de pesos actual
        Salida:
            costo: valor del costo utilizando los pesos w de la entrada
            gradiente: gradiente del costo respecto a los pesos de entrada w. 
        '''
           
        
        N = X.shape[0]
        

        
        # se inicializa el costo
        costo = 0.
        
        # se inicializa la variable que guarda el gradiente de la época t
        gradiente = np.zeros(X.shape[1]) 
                
        for i in range(N):
            
            #################################################################################
            #################  EMPIEZA ESPACIO PARA COMPLETAR CÓDIGO ########################
            #################################################################################
            
            # 1. se actualiza el gradiente
            gradiente += -y[i]*X[i]*sigmoide(-y[i]*(X[i]@w)) 
            
            # 2. se actualiza el costo, se sugiere usar la función sigmoide implementada más arriba
            # para evitar posibles problemas numéricos
            costo += np.log(1+np.exp(-y[i]*(w @ X[i])))
            
            #################################################################################
            #################  TERMINA ESPACIO PARA COMPLETAR CÓDIGO ########################
            #################################################################################
       
        
        gradiente /= N
        costo /= N
        
        return costo, gradiente
```

```python
# Se testea calcular_costo_y_gradiente_sin_regularizar()
np.random.seed(33)
X5 = np.random.randn(3,4)
X5[:,0]=1
y5 = np.array([1,-1,1,-1])
w5 = np.random.randn(4)
c5, g5 = calcular_costo_y_gradiente_sin_regularizar(X5,y5,w5)
assert np.allclose(c5, 0.13048058961396589), 'Revisar el cálculo del costo'
g5_ = np.array([-0.10645973,  0.10632848,  0.12291721,  0.07835299])
assert np.allclose(g5, g5_), 'Revisar el cálculo del gradiente'
```

### Ejercicio 3 - parte e)  

Completar la implementación del método `regresión_logistica()` minimizando la función de costo mediante el método de *batch gradient descent*. Como condición de parada se utilizará solamente el número de iteraciones.

```python
def regresion_logistica(X, y, lr = 0.1, max_iter=1000, 
                        w_inicial=None, reg_factor=0):
    '''
    Entrada:
        X: matriz de tamaño Nx(d+1) donde cada fila representa una muestra
        y: vector de tamaño N que contiene las etiquetas
        lr: learning rate
        max_iter: si el número de iteraciones alcanza a max_iter se detiene la optimización.
    Salida:
        w: vector de pesos encontrado
        costo: arreglo de tamaño igual al número de iteraciones que almacena los valores 
               que toma la función de costo         
    '''     
    
    if w_inicial is None:
        # se inicializan los pesos a valores aleatorios muy pequeños
        w = 1e-6 * np.random.randn(X.shape[1])
    else:
        w = w_inicial.copy()
    costo = []    
    
    t=0 # contador de iteraciones

    while (t < max_iter):
        
        #################################################################################
        #################  EMPIEZA ESPACIO PARA COMPLETAR CÓDIGO ########################
        #################################################################################
        
        # 1. se calcula el costo y el gradiente del costo respecto a los pesos actuales
        costo_sin_reg, grad_sin_reg = calcular_costo_y_gradiente_sin_regularizar(X, y, w) 
        
        # 2. se agrega el término de regularización al gradiente
        grad_t = grad_sin_reg + reg_factor*w
        
        # 3 se agrega el termino de regularización al costo
        costo_t = costo_sin_reg + (reg_factor/2)*np.sum(w**2)
        
        # 4. se agrega el costo correspondiente a esta iteración a la lista
        costo.append(costo_t)
        
        # 5. se actualiza el vector de pesos w
        w = w - lr*grad_t
       
        # 6. se incrementa en 1 el valor del contador de iteraciones
        t += 1

       #################################################################################
       #################  TERMINA ESPACIO PARA COMPLETAR CÓDIGO ########################
       #################################################################################
    
    
    return w, costo
```

### Ejercicio 3 - parte g)

```python
def accuracy(X, y, w):
    '''
    Entrada:
       X: arreglo de (N, d+1) con las características expresadas en coordenadas homogéneas
       y: arreglo de tamaño N con las etiquetas
       w: vector de pesos del clasificador
    Salida:
       accuracy: tasa de acierto  (valor entre 0 y 1)
    '''
    #################################################################################
    #################  EMPIEZA ESPACIO PARA COMPLETAR CÓDIGO ########################
    #################################################################################
    
    punto_funcionamiento = 0.50
    
    P_spam = sigmoide(X@w)
    
    y_pred = 2*(P_spam > punto_funcionamiento) - 1
    
    accuracy = np.average(y_pred==y)
    
    #################################################################################
    #################  TERMINA ESPACIO PARA COMPLETAR CÓDIGO ########################
    #################################################################################
    
    return accuracy
```
